## 질문 2 : 기업에서 원하는 툴

```python
tools = df1['description_tokens']
tools.head(10)
```

```python
def tokenss(tool):
    tool_lst=[]
    a= []
    for i in tool:
        a = i.replace('[','').replace(']','').replace("'",'').replace(' ','')
        if len(a)!=0:
            for j in a.split(','):
                tool_lst.append(j)
    return tool_lst
        
```

```python
tool_lst = tokenss(tools)
tool_lst
```

```python
df_tool = pd.DataFrame(tool_lst)
```

```python
tool_lst.count('pl/sql')
```

```python
df_tool.rename(columns={0:'Tool'},inplace=True)
```

```python
df_tool
```

```python
df_tool = df_tool.replace('pl/sql','sql')
df_tool['Tool'].unique()
```

```python
df1_tool = df_tool.groupby('Tool')['Tool'].count().sort_values(ascending=False)
df1_tool
```

```python
import plotly.graph_objects as go
fig = go.Figure()
fig.add_trace(
    go.Bar(
        x=df1_tool.head(15).index, y=df1_tool.head(15).values, text=df1_tool.head(15).values, textposition='auto', texttemplate='%{y}'
    )
)

fig.update_layout(
    {
        "title": {
            "text": "Tools used in practice",
            "x": 0.5,
            "y": 0.9,
            "font": {
                "size": 20
            }
        },
        # "showlegend": True,
        "xaxis": {
            "title": "Tools",
            "showticklabels":True,
            "dtick": 1
            },
        "yaxis": {
            "title": "Count"
        },
        "autosize":False,
        "width":1000,
        "height":500
    }
)
fig.show()
```

```python
df_salary = df.dropna(subset='salary_pay')
df_salary
```

```python
df_salary['salary_pay'].str.split('–')
```

```python
df_salary[df_salary['salary_pay'].str.contains('K')]
```

```python
import numpy as np
income_mean = []
for i in df_salary['salary_pay'].str.split('–'):
    income = []
    # print(i)
    # print(type(i))
    for j in i:
        # print(j)
        if 'K' in j:
            income.append(float(j.replace('K',''))*1000)
        else:
            income.append(float(j))
    income_mean.append(np.mean(income))
            
```

```python
df_salary['income_mean'] = income_mean
```
