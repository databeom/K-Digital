# 데이터 불러오기
- 1949 ~ 2010년 데이터 셋 : population_1
- 2015 ~ 2022년 데이터 셋 : population_2

```python
import pandas as pd
population_1 = pd.read_csv('population_1.csv',encoding='cp949')
population_1
```

```python
population_2 = pd.read_csv('population_2.csv',encoding='cp949')
population_2.head()
```

## 데이터 전처리

### 필요없는 coulmn, row 삭제

```python
# 필요없는 column 삭제
population_1.drop(['연령별'],axis=1,inplace=True)
population_1.head()
```

```python
# 필요없는 row 삭제
population_1 = population_1[1:]
```

### 필요없는 문자열('-') 0으로 대체

```python
population_1 = population_1.replace('-','0')
```

### 데이터 타입 변경

```python
population_1.set_index('행정구역별',inplace=True)
population_1 = population_1.astype('int')
population_1.reset_index(inplace=True)
population_1.info()
```

### 행정구역 명 통일

```python
population_1.replace({
    '서울특별시':'서울',
    '부산광역시':'부산',
    '대구광역시':'대구',
    '인천광역시':'인천',
    '광주광역시':'광주',
    '대전광역시':'대전',
    '울산광역시':'울산',
    '경기도':'경기',
    '강원도':'강원',
    '충청북도':'충북',
    '충청남도':'충남',
    '전라북도':'전북',
    '전라남도':'전남',
    '경상북도':'경북',
    '경상남도':'경남',
    '제주도':'제주'
},inplace=True)
```

```python
population_2.replace({
    '서울특별시':'서울',
    '부산광역시':'부산',
    '대구광역시':'대구',
    '인천광역시':'인천',
    '광주광역시':'광주',
    '대전광역시':'대전',
    '울산광역시':'울산',
    '경기도':'경기',
    '강원도':'강원',
    '충청북도':'충북',
    '충청남도':'충남',
    '전라북도':'전북',
    '전라남도':'전남',
    '경상북도':'경북',
    '경상남도':'경남',
    '제주특별자치도':'제주'
},inplace=True)
```

### population_2 컬럼명 변경
- population_1과 population_2행정구역에 해당하는 컬럼명이 다르므로 통일 시켜줘야함

```python
population_2.rename(columns ={'행정구역별(읍면동)':'행정구역별'},inplace=True)
```

### 데이터 병합
- 병합한 데이터 셋 : population

```python
population = pd.merge(population_1,population_2,how='left',on='행정구역별')
```

### 인구 비율 데이터 셋 만들기

```python
population_per = population.copy()
for col in population_per.columns[1:]:
    population_per[col] = population_per[col].apply(lambda x: (x/population_per[col].sum())*100)
```

## 데이터 내보내기

```python
population.to_csv("population.csv")
population_per.to_csv("population_per.csv")
```

## 최종 코드

```python
import pandas as pd

# 데이터 불러오기
population_1 = pd.read_csv('korea_population.csv',encoding='cp949')
population_2 = pd.read_csv('korea_population_2022.csv',encoding='cp949')

#데이터 전처리

## 필요없는 column 삭제
population_1.drop(['연령별'],axis=1,inplace=True)

## 필요없는 row 삭제
population_1 = population_1[1:]

##문자열 0으로 대체
population_1 = population_1.replace('-','0')

## 데이터 타입 변경
population_1.set_index('행정구역별',inplace=True)
population_1 = population_1.astype('int')
population_1.reset_index(inplace=True)

## 행정구역 명 통일
population_1.replace({
    '서울특별시':'서울',
    '부산광역시':'부산',
    '대구광역시':'대구',
    '인천광역시':'인천',
    '광주광역시':'광주',
    '대전광역시':'대전',
    '울산광역시':'울산',
    '경기도':'경기',
    '강원도':'강원',
    '충청북도':'충북',
    '충청남도':'충남',
    '전라북도':'전북',
    '전라남도':'전남',
    '경상북도':'경북',
    '경상남도':'경남',
    '제주도':'제주'
},inplace=True)

population_2.replace({
    '서울특별시':'서울',
    '부산광역시':'부산',
    '대구광역시':'대구',
    '인천광역시':'인천',
    '광주광역시':'광주',
    '대전광역시':'대전',
    '울산광역시':'울산',
    '경기도':'경기',
    '강원도':'강원',
    '충청북도':'충북',
    '충청남도':'충남',
    '전라북도':'전북',
    '전라남도':'전남',
    '경상북도':'경북',
    '경상남도':'경남',
    '제주특별자치도':'제주'
},inplace=True)

## population_2 컬럼명 변경
population_2.rename(columns ={'행정구역별(읍면동)':'행정구역별'},inplace=True)

## 데이터 병합
population = pd.merge(population_1,population_2,how='left',on='행정구역별')

## 인구 비율 데이터 셋 만들기
population_per = population.copy()
for col in population_per.columns[1:]:
    population_per[col] = population_per[col].apply(lambda x: (x/population_per[col].sum())*100)

## 데이터 내보내기
population.to_csv("population.csv")
population_per.to_csv("population_per.csv")
```
